```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "Age":[20,30,35,40],
    "Salary": [2000,3000,4000,10000]
}
df = pd.DataFrame(data)
```

```python
df.plot(x='Age',y='Salary',kind="line",marker=".")
plt.title("Salary Distribution")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
df.plot(x='Age',y='Salary',kind="bar",color='black')
plt.title("Salary Distribution")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
import numpy as np
```

```python
np.random.seed(0)
data = pd.date_range(start="2023-01-01",end="2023-12-31",freq="D")
sales = np.random.normal(loc=1000,scale=300,size=len(data))
df = pd.DataFrame({'Sales':sales,"Date" : data})
```

```python
df['Month'] = df['Date'].dt.to_period("M")
```

```python
df
```

```python
monthly_sales = df.groupby("Month")['Sales'].sum()
```

```python
plt.plot(monthly_sales.index.astype(str),monthly_sales.values,marker="x",linestyle="-")
```
