```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
url = 'https://www.daraz.com.np/gaming-consoles/?spm=a2a0e.11779170.cate_5.7.287d2d2bHiXucD'
webpage = requests.get(url,headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
vi = soup.find('div',class_='ant-col-19 ant-col-push-5 side-right--Tyehf')
```

```python
divs = soup.find('body')
```

```python
yo = divs.find_all('li')
```

```python
yo
```

```python
data = soup.find_all('h6',class_='q_title')
movie = soup.find_all('p',class_='Directors')
Producer = soup.find_all('p',class_='Producer')
gen = soup.find_all('p',class_='Genre')
Edi = soup.find_all('p',class_='Editor')
Names= soup.find_all('p',class_='')
```

```python
Movie=[]
Dialogue = []
Directors = []
Producers = []
Genre = []
Editor = []
for i in range(len(gen)):
    Movie.append(Names[i].text)
    Dialogue.append(data[i].text.replace('{}.'.format(i+1),''))
    Directors.append(movie[i].text.replace('Directors:',''))
    Producers.append(Producer[i].text.replace('Producer:',''))
    Genre.append(gen[i].text.replace('Genre:',''))
    Editor.append(Edi[i].text.replace('Editor:',''))
```

```python
dicto = {'Movie':Movie,'Dialogue':Dialogue,'Directors':Directors,'Producers':Producers,'Genre':Genre,'Editor':Editor}
```

```python
dataframe = pd.DataFrame(dicto)
```
