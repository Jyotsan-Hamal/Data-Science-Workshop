## EDA ('Exploratory Data Analysis')

### 1.Data Visualization - Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
x = [1,2,3,5,6,8]
y = [1,3,4,5,10,13]
plt.plot(x,y,linewidth=0.5)
plt.axis((0,15,0,15))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Simple Plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [1,4,6,8,10]
y2 = [3,6,8,14,19]
plt.plot(x,y1,label='Line 1', color='red',linestyle='--')
plt.plot(x,y2,label='Line 2', color='blue',linestyle='-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple lines plot")
plt.show()
```

## Bar Graph

```python
categorical = ['smal','medium','large','XL']
values = [20,25,30,35]
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar plot")
plt.bar(categorical,values,color='green')
```

```python
x = [1,2,3,4,5]
y1 = [1,4,6,8,10]
y2 = [3,6,8,14,19]
plt.plot(x,y1,'bo',label='Line 1', color='red',linestyle='--')
plt.plot(x,y2,label='Line 2', color='blue',linestyle='-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple lines plot")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,6,8,10]
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter plot")
plt.show()
```

```python
names = ['group_a','group_b','group_c']
values=[1,10,100]
plt.figure(figsize=(10,4))

plt.subplot(1,3,1) # (row count,column,which one is this)
plt.bar(names,values,color='red')

plt.subplot(1,3,2)# (row,column,which one is this)
plt.scatter(names,values,color='blue')

plt.subplot(1,3,3)# (row,column,which one is this)
plt.plot(names,values,color='green')
plt.suptitle("Categorical Plotting")
plt.show()
```

```python
names = ['group_a','group_b','group_c']
values=[1,10,100]
plt.figure(figsize=(10,4))

plt.subplot(2,2,1) # (row count,column,which one is this)
plt.bar(names,values,color='red')

plt.subplot(2,2,3)# (row,column,which one is this)
plt.scatter(names,values,color='green')

plt.subplot(1,3,3)# (row,column,which one is this)
plt.plot(names,values)
plt.suptitle("Categorical Plotting")
plt.show()
```

```python
# Create some fake data.
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('A tale of 2 subplots')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Damped oscillation')

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('Undamped')

plt.show()
```
